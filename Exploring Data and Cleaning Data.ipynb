```python
## Z-Score method
upper_limit = read_data_subset_copy['semantic_scholar'].mean() + 3* read_data_subset_copy['semantic_scholar'].std() 
lower_limit = read_data_subset_copy['semantic_scholar'].mean() - 3* read_data_subset_copy['semantic_scholar'].std() 
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
#find the outliers
read_data_subset_copy.loc[(read_data_subset_copy['semantic_scholar'] > upper_limit) | (read_data_subset_copy['semantic_scholar'] < lower_limit)]
```

```python
# trimming - delete the outlier data
new_read_data_subset_copy = read_data_subset_copy.loc[(read_data_subset_copy['semantic_scholar'] < upper_limit) & (read_data_subset_copy['semantic_scholar'] > lower_limit)]
print('Before removing outliers:', len(read_data_subset_copy))
print('After removing outliers:', len(new_read_data_subset_copy))
print('outliers removed:',len(read_data_subset_copy)- len(new_read_data_subset_copy))
```

```python
sns.boxplot(new_read_data_subset_copy['semantic_scholar'])
```

```python
#capping - change the outlier values to upper or lower limit values
new_read_data_subset_copy = read_data_subset_copy.copy()
new_read_data_subset_copy.loc[new_read_data_subset_copy['semantic_scholar']>upper_limit, 'semantic_scholar']=upper_limit
new_read_data_subset_copy.loc[new_read_data_subset_copy['semantic_scholar']<lower_limit, 'semantic_scholar']=lower_limit
```
